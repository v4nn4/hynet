```python
from hynet.prepare import generate_character_image
from hynet.evaluate import single_evaluate
from PIL import ImageFont, Image
import numpy as np


N = 32
characters = "ԱԲԳԴԵԶԷԸԹԺԻԼԽԾԿՀՁՂՃՄՅՆՇՈՉՊՋՌՍՎՏՐՑՒՓՔՕՖաբգդեզէըթժիլխծկհձղճմյնշոչպջռսվտրցւփքօֆ"
for i, c in enumerate(characters):
    font = ImageFont.truetype("hynet/fonts/Mk_Parz_U-Italic.ttf", N)
    character = generate_character_image(c, font, N)
    Image.fromarray((255 * character[0].numpy()).astype(np.uint8), mode="L").save(f"build/logs/evaluate/character_{i}.png")
    print(c, single_evaluate(character, N))
```
